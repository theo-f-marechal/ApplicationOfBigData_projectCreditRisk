We also filter out absurd values for numerical features.

```python
df_trimed[['DAYS_BIRTH', 
           'DAYS_EMPLOYED', 
           'DAYS_REGISTRATION', 
           'DAYS_ID_PUBLISH']] = 
    df_trimed.loc[:,['DAYS_BIRTH', 
                     'DAYS_EMPLOYED', 
                     'DAYS_REGISTRATION', 
                     'DAYS_ID_PUBLISH']] *  -1
df_trimed
```
